```python
# Read the data and create the dependent and independent variable arrays
dataset = pd.read_csv('Position_Salaries.csv')
x=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
x=sc_x.fit_transform(x)
y=sc_y.fit_transform(y)
```

```python
dataset.plot('Level','Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
predl=reg.predict(x)
```

```python
from sklearn.svm import SVR
svreg=SVR(kernel='rbf')
svreg.fit(x,y)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,predl,color="blue")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
predp=svreg.predict(x)
plt.scatter(x,y,color="red")
#plt.legend("Actual")
plt.plot(x,predp,color="blue")
#plt.legend("Poly")
#plt.legend("Linear")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title ("Truth Or Bluff")
plt.legend(loc=2)
plt.show()
```

```python
# Polynomial features automatically adds ones for X0 term in the multiple linear regression

from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)
xpoly=pf.fit_transform(x)
```
